```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

Generate a list of columns interest in

```python
rng = [i for i in range(1,50)]
#rng
```

```python
df = pd.read_csv("PowerUsage.csv", skiprows=1, header=None, usecols=rng, parse_dates=[1], index_col=0 )
df.head(2)
```

Set index name to have a proper index name

```python
df.index.names = ['Date']
df.head(2)
```

Generate a list of half an hour date range for the column heading

```python
# Generate Time interval
rng = pd.date_range(start='00:00', end='23:59', freq='30min').strftime('%H:%M')
rng### Set Date to be the index of the dataframe
```

Assign name to each of the column

```python
df.columns = rng
df.head(2)
```

Display data for some date

```python
df.loc['2019-01-01':'2019-01-04']
```

Plot a graph 

```python
df.loc['2019-05-20', list(pd.date_range(start='05:00', end='23:30', freq='30min').strftime('%H:%M'))].astype(float).plot()
```

From 11:00 PM to 7:00 AM are Off peak time

```python
offpeak= list(pd.date_range(start='00:00', end='6:30', freq='30min').strftime('%H:%M')) + ['23:00', '23:30']
offpeak
```

```python
peak = [item for item in list(rng) if item not in offpeak]
peak
```

```python
df['total'] = df.iloc[:].sum(axis=1)
df['offpeak'] =df.iloc[:, df.columns.isin(offpeak)].sum(axis=1)
df['peak'] =df.iloc[:, df.columns.isin(peak)].sum(axis=1)
```

Resample to monthly

```python
df['total'].resample('M').sum()
```
